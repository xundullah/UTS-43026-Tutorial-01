### **Exercise #03: Market Equilibrium and Surplus Calculation**  

Economists estimate that the following equation defines the **supply function** for the widget market:

\[
q = 0.2\pi - 40
\]

where:
- $ q $ represents **quantity supplied**.
- $ \pi $ represents **the unit price ($/unit)**.

We will determine:
1. **The equilibrium price and quantity** (where demand equals supply).
2. **Consumer surplus (CS)**.
3. **Producer surplus (PS)**.
4. **Total surplus (TS = CS + PS)**.

---

## **(a) Calculate the Demand and the Price at Market Equilibrium**  

### **Step 1: Given Demand and Supply Functions**  
From **Exercise #02**, the **inverse demand function** is:

\[
\pi = -10q + 2000
\]

The **supply function** is:

\[
q = 0.2\pi - 40
\]

At **market equilibrium**, the **quantity demanded equals the quantity supplied**, meaning both functions should be equal.

---

### **Step 2: Express $ \pi $ in Terms of $ q $ for the Supply Function**  
Rearrange the supply function to express $ \pi $ in terms of $ q $:

\[
q + 40 = 0.2\pi
\]

\[
\pi = 5q + 200
\]

---

### **Step 3: Solve for Equilibrium Quantity**  
At equilibrium, the **demand function** equals the **supply function**:

\[
-10q + 2000 = 5q + 200
\]

Rearrange the equation:

\[
2000 - 200 = 10q + 5q
\]

\[
1800 = 15q
\]

\[
q^* = \frac{1800}{15} = 120
\]

Thus, the **equilibrium quantity** is:

\[
q^* = 120 \text{ units}
\]

---

### **Step 4: Solve for Equilibrium Price**  
Substituting $ q = 120 $ into the **supply function**:

\[
\pi^* = 5(120) + 200
\]

\[
= 600 + 200
\]

\[
= 800 \text{ \$/unit}
\]

Thus, the **equilibrium price** is:

\[
\pi^* = 800 \text{ \$/unit}
\]

---

## **(b) Calculate Consumer Surplus, Producer Surplus, and Total Surplus**  

### **Step 1: Consumer Surplus (CS) Calculation**
Consumer surplus is the area of the triangle between:
- The **maximum price consumers are willing to pay** ($ \pi_{\max} $).
- The **equilibrium price** ($ \pi^* $).
- The **equilibrium quantity** ($ q^* $).

\[
\text{Consumer Surplus} = \frac{1}{2} \times q^* \times (\pi_{\max} - \pi^*)
\]

From previous calculations:
- $ \pi_{\max} = 2000 $ (price at $ q = 0 $).
- $ \pi^* = 800 $.
- $ q^* = 120 $.

Substituting the values:

\[
\text{CS} = \frac{1}{2} \times 120 \times (2000 - 800)
\]

\[
= \frac{1}{2} \times 120 \times 1200
\]

\[
= \frac{144000}{2}
\]

\[
= 72,000
\]

Thus, the **consumer surplus** is:

\[
CS = 72,000 \text{ \$/unit}
\]

---

### **Step 2: Producer Surplus (PS) Calculation**
Producer surplus is the area of the triangle between:
- The **equilibrium price** ($ \pi^* $).
- The **minimum price producers are willing to accept** ($ \pi_{\min} $).
- The **equilibrium quantity** ($ q^* $).

\[
\text{Producer Surplus} = \frac{1}{2} \times q^* \times (\pi^* - \pi_{\min})
\]

From the **supply function**:
- $ \pi_{\min} = 200 $ (price at $ q = 0 $).
- $ \pi^* = 800 $.
- $ q^* = 120 $.

Substituting the values:

\[
\text{PS} = \frac{1}{2} \times 120 \times (800 - 200)
\]

\[
= \frac{1}{2} \times 120 \times 600
\]

\[
= \frac{72000}{2}
\]

\[
= 36,000
\]

Thus, the **producer surplus** is:

\[
PS = 36,000 \text{ \$/unit}
\]

---

### **Step 3: Total Surplus (TS) Calculation**
Total surplus is the sum of **consumer surplus** and **producer surplus**:

\[
\text{Total Surplus} = CS + PS
\]

\[
= 72,000 + 36,000
\]

\[
= 108,000
\]

Thus, the **total surplus** is:

\[
TS = 108,000 \text{ \$/unit}
\]

---

## **Final Summary**
| **Calculation** | **Value** |
|---------------|-----------|
| **Equilibrium Quantity ($ q^* $)** | $ 120 $ units |
| **Equilibrium Price ($ \pi^* $)** | $ 800 $ \$/unit |
| **Consumer Surplus (CS)** | $ 72,000 $ \$/unit |
| **Producer Surplus (PS)** | $ 36,000 $ \$/unit |
| **Total Surplus (TS)** | $ 108,000 $ \$/unit |

---

## **Key Takeaways**
1. **Market equilibrium occurs where demand equals supply**, resulting in:
   - **120 units sold** at **\$800/unit**.
2. **Consumer Surplus (CS) measures the total benefit consumers receive**, calculated as:
   - **$ 72,000 $ \$/unit**.
3. **Producer Surplus (PS) measures the total benefit producers receive**, calculated as:
   - **$ 36,000 $ \$/unit**.
4. **Total Surplus (TS = CS + PS) represents the total economic benefit from trade**:
   - **$ 108,000 $ \$/unit**.
5. **Market efficiency is maximized at equilibrium**, ensuring both consumer and producer benefits.
