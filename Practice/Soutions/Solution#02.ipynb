### **Problem #02: Inverse Demand Function Analysis**  

The inverse demand function of a group of consumers for a given type of widgets is given by:

$$
\pi = -8q + 2500 \quad [\$/\text{unit}]
$$

where:
- $ q $ represents **demand (quantity of widgets purchased)**.
- $ \pi $ represents **the unit price ($/unit) of the widgets**.

We will determine:
1. **The maximum consumption of these consumers**.
2. **The price that no consumer is prepared to pay for this product**.
3. **The maximum consumers’ surplus** and why it cannot be realized.
4. **The change in consumption and revenue when price increases by 25%**.

---

## **(a) Determine the Maximum Consumption of These Consumers**  

### **Step 1: Understanding Maximum Consumption**  
- The maximum number of widgets that can be **purchased by consumers** occurs when the **price $ \pi = 0 $**.
- This means consumers are willing to **consume the maximum possible quantity when the product is free**.

### **Step 2: Solve for $ q_{\max} $**  
- Set $ \pi = 0 $ in the given demand equation:

  $$
  0 = -8q + 2500
  $$

- Solve for $ q $:

  $$
  q_{\max} = \frac{2500}{8} = 312.5 \text{ units}
  $$

- Since fractional demand doesn’t make sense in this context, we round it to **312 units**.

### **Step 3: Conclusion**  
- The **maximum consumption** is:

  $$
  q_{\max} = 312 \text{ units}
  $$

---

## **(b) Determine the Price That No Consumer Is Willing to Pay**  

### **Step 1: Understanding the Maximum Price**  
- The maximum price consumers are willing to pay is when **demand is zero** ($ q = 0 $).
- At this price, **no widgets are purchased**.

### **Step 2: Solve for $ \pi_{\max} $**  
- Set $ q = 0 $ in the inverse demand function:

  $$
  \pi = -8(0) + 2500
  $$

  $$
  \pi_{\max} = 2500 \text{ \$/unit}
  $$

### **Step 3: Conclusion**  
- The price that **no consumer is prepared to pay** is:

  $$
  \pi_{\max} = 2500 \text{ \$/unit}
  $$

---

## **(c) Determine the Maximum Consumers’ Surplus and Explain Why It Is Not Realizable**  

### **Step 1: Understanding Consumers' Surplus**  
- **Consumer surplus** represents the **total benefit** consumers receive when they **pay less than their maximum willingness to pay**.
- It is the **area under the inverse demand curve**, which is a **triangle** in this case.

### **Step 2: Consumer Surplus Formula**  
- The **maximum consumer surplus** occurs when:

  - **$ q = q_{\max} $** (maximum consumption).
  - **$ \pi = 0 $** (product is free).
  - The formula for consumer surplus:

    $$
    \text{Consumer Surplus} = \frac{q_{\max} \times \pi_{\max}}{2}
    $$

### **Step 3: Calculate Consumer Surplus**  
- Substituting the values:

  $$
  \text{Consumer Surplus} = \frac{312 \times 2500}{2}
  $$

  $$
  = 390,000 \text{ \$/unit}
  $$

### **Step 4: Why Is This Surplus Not Realizable?**  
- **Producers will not sell for $ \pi = 0 $** because:
  - No rational business sells at **zero price**.
  - A **profit-driven firm** must sell at a price that **covers costs** and generates revenue.

### **Step 5: Conclusion**  
- The **maximum consumer surplus** is **$ 390,000 $**, but it **cannot be achieved** because **producers will not give products away for free**.

---

## **(d) If the Price Increases by 25%, Calculate the Change in Consumption and Revenue**  

### **Step 1: Correct Initial Price**  
- The price when **$ q = 100 $** was **$ \pi = 1700 $** (calculated from the demand equation).
- A **25% price increase** means the new price is:

  $$
  \pi' = 1700 \times (1 + 0.25)
  $$

  $$
  = 2125 \text{ \$/unit}
  $$

### **Step 2: Solve for New Consumption Level**  
- Using the inverse demand function:

  $$
  q = \frac{\pi - 2500}{-8}
  $$

- Substituting $ \pi' = 2125 $:

  $$
  q' = \frac{2125 - 2500}{-8} = \frac{-375}{-8}
  $$

  $$
  = 46.875 \approx 47 \text{ units}
  $$

### **Step 3: Calculate Change in Consumption**  
- **Initial consumption** at $ \pi = 1700 $:

  $$
  q_{\text{old}} = 100
  $$

- **Change in consumption**:

  $$
  \Delta q = q' - q_{\text{old}} = 47 - 100 = -53 \text{ units}
  $$

### **Step 4: Calculate Revenue Before and After**  
- **Old Revenue**:

  $$
  R_{\text{old}} = 100 \times 1700 = 170,000 \text{ \$/unit}
  $$

- **New Revenue**:

  $$
  R_{\text{new}} = 47 \times 2125 = 99,875 \text{ \$/unit}
  $$

### **Step 5: Calculate Change in Revenue**  
- **Change in revenue**:

  $$
  \Delta R = R_{\text{new}} - R_{\text{old}}
  $$

  $$
  = 99,875 - 170,000
  $$

  $$
  = -70,125 \text{ \$/unit}
  $$

### **Step 6: Conclusion**  
- **Change in consumption**: **$ -53 $ units** (53 fewer units sold).
- **Change in revenue**: **$ -70,125 $ \$/unit**.
